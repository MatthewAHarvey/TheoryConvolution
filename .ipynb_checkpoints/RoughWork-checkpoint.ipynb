```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
from scipy.interpolate import interp1d
```

### Convolving the theory data with a gaussian (FWHM = $\pm 3^\circ$)

All of the cross-section theory data from Don Madison's group is assuming a perfect spectrometer with $0^\circ$ angular resolution. In our (e,2e) experiment, the detectors have an acceptance angle sensitivity in the form of a gaussian with FWHM of about $6^\circ$. In order to more properly compare the theory and experimental results, the theory data should be convolved with an appropriate gaussian weighting.

First step is load in the data from the file to be analysed

```python
file_30eV = '30eV theory 10째 to 36째.txt'
file_50eV = '50eV theory 10째 to 36째.txt'

with open(file_30eV, 'r', encoding = 'mac-roman') as f:
    contents = f.readlines()
    header = contents[0].split('\t')[:-1]
```

Now obtain the swept angles and the data in the form of an array. Then plot the data for each fixed angle to get an idea of what we are looking at.

```python
# Obtain the fixed angles as a numpy array
fixedAngles = np.zeros(len(header) - 1)
for i, angle in enumerate(header[1:]):
    fixedAngles[i] = float(angle.split('째')[0])

# Obtain swept angles as a numpy array
sweptAngles = np.zeros(15)
data = np.nan * np.ones((15, 17)) # rows, columns
for i, line in enumerate(contents[1:16]):
    lineList = line.split('\t')[:-1]
    sweptAngles[i] = lineList[0]
    data[i, :] = lineList[1:]

# sweptAngles
# data[:,1] # 14 deg data
def dressFigure():
    plt.xlabel(r'Analyser Angle $\theta_2$ ($^{\circ}$)')
    plt.ylabel(r'Counts')
    fig = plt.gcf()
    ax = plt.gca()
    xmajorLocator = MultipleLocator(10)
    xminorLocator = MultipleLocator(5)
    ax.xaxis.set_major_locator(xmajorLocator)
    ax.xaxis.set_minor_locator(xminorLocator)
    plt.tick_params(which = 'both', direction = 'in', left='on', top='on', right='on', bottom='on')
    plt.grid()
    ax.grid(linestyle='-', linewidth='0.5', color='gray', alpha=0.2)
    plt.legend()
    plt.show()
    
for i, counts in enumerate(data.T):
    plt.figure(figsize = (8,5), dpi = 276/1.25)
    plt.plot(sweptAngles, counts, '.', label = header[i+1])
    dressFigure()
```

So now we have the data and can plot it but it isn't very detailed and doesn't lead the eye. Plot it again but this time with a [cubic spline fit](https://docs.scipy.org/doc/scipy/reference/tutorial/interpolate.html).

```python
interpSpacing = 0.1
interpAngles = np.arange(0,140 + interpSpacing, interpSpacing)
for i, counts in enumerate(data.T):
    cubicSplineForm = interp1d(sweptAngles, counts, kind='cubic')
    # interpAngles = np.linspace(0, 140, 501)
    splineData = cubicSplineForm(interpAngles)
    plt.figure(figsize = (8,5), dpi = 276/1.25)
    plt.plot(sweptAngles, counts, '.', label = header[i+1])
    plt.plot(interpAngles, splineData, label = 'cubic spline')
    dressFigure()
```

That worked really well and the fit looks excellent.

Since our detectors will be blurring this signal out, maybe we should convolve the swept angle data as well and compare this to our experimental results.

In fact, why not just do a cubic spline interpolation over the 2D data and then Gaussian blur that with FWHM = 6$^\circ$ in both axes? Then just plot the slices through the angles we want?

Try it... do cubic interpolation on one axis and then the other:

```python
# print(f'Swept Angle Length: {len(sweptAngles)}')
# print(f'Fixed Angle Length: {len(fixedAngles)}')
# print(data.shape)

# Produce the angle arrays for the interpolation grid
interpSweptAngles = np.arange(0, 140 + interpSpacing, interpSpacing)
interpFixedAngles = np.arange(10, 36 + interpSpacing, interpSpacing)
# print(f'Interpolated Swept Angle Length: {len(interpSweptAngles)}')
# print(f'Interpolated Fixed Angle Length: {len(interpFixedAngles)}')
# fixedAngleInterpolatedData = np.zeros((len(sweptAngles), len(interpFixedAngles)))

# first interpolate over the swept angles as before
sweptAngleInterpolatedData = np.zeros((len(interpSweptAngles), len(fixedAngles)))
# print(sweptAngleInterpolatedData.shape)
for i, counts in enumerate(data.T): # iterate over the fixed angles
    f = interp1d(sweptAngles, counts, kind = 'cubic')
    sweptAngleInterpolatedData[:,i] = f(interpSweptAngles)

# now interpolate over the fixed angles
interpolatedData = np.zeros((len(interpSweptAngles), len(interpFixedAngles)))
# print(interpolatedData.shape)
for i, counts in enumerate(sweptAngleInterpolatedData):
    f = interp1d(fixedAngles, counts, kind = 'cubic')
    interpolatedData[i,:] = f(interpFixedAngles)

del sweptAngleInterpolatedData # don't need this anymore

# Now compare the non interpolated theory data with the interpolated stuff to see if it looks right
plt.figure(dpi=276)
plt.xlabel('Swept Angles ($^\circ$)')
plt.ylabel('Fixed Angles ($^\circ$)')
plt.imshow(data.T, extent=[0,140,36,10])
ax = plt.gca()
ax.set_aspect(5)
plt.show()

# interpolatedData /= interpolatedData.max()
plt.figure(dpi=276)
plt.xlabel('Swept Angles ($^\circ$)')
plt.ylabel('Fixed Angles ($^\circ$)')
plt.imshow(interpolatedData.T, extent=[0,140,36,10])
ax = plt.gca()
ax.set_aspect(5)
plt.show()
```

```python
interpolatedData.shape
```

That looks good. There is a little banding around 22.5$^\circ$ but it's probably not too bad.

Now need to do a gaussian filter on it with FWHM = 6$^\circ$. Can do this by convolving with a gaussian. Do it one point at a time or in frequency domain?

First let's just try using numpy's built in convolve function. Test by convolving a top hat with a gaussian:

```python
xlimit = 8
x = np.arange(-xlimit, xlimit + interpSpacing, interpSpacing)
tophat = np.zeros(interpolatedData.shape[1])
tophat[110:115] = 1
plt.plot(tophat)
plt.title('Top Hat')
plt.show()

xcentre = 0
sigma = 6/2.355
gaussian = np.exp(-(x-xcentre)**2/(2*sigma**2))
plt.title(f'Gaussian with $\sigma =$ {sigma:.2f} $^\circ$')
plt.plot(gaussian)
plt.show()

conv = np.convolve(tophat, gaussian, mode = 'same')
plt.plot(conv)
plt.title('Convolved Gaussian/Top Hat')
plt.show()
```

Going out to $\pm13$ seems to have just about enough points. Now convolve across both axes.

```python
convolvedData = np.zeros(interpolatedData.shape)
for i, counts in enumerate(interpolatedData):
    convolvedData[i,:] = np.convolve(counts, gaussian, mode = 'same')
for i, counts in enumerate(interpolatedData.T):
    convolvedData[:,i] = np.convolve(counts, gaussian, mode = 'same')

# renormalise to previous max
convolvedData /= convolvedData.max()
convolvedData *= interpolatedData.max()

plt.figure(dpi=276)
plt.xlabel('Swept Angles ($^\circ$)')
plt.ylabel('Fixed Angles ($^\circ$)')
plt.imshow(convolvedData.T, extent=[0,140,36,10])
ax = plt.gca()
ax.set_aspect(5)
plt.show()

# plot the difference between the convolved and normal data.
plt.figure(dpi=276)
plt.title('Difference between convolved and original theory')
plt.xlabel('Swept Angles ($^\circ$)')
plt.ylabel('Fixed Angles ($^\circ$)')
plt.imshow(convolvedData.T - interpolatedData.T, extent=[0,140,36,10])
ax = plt.gca()
ax.set_aspect(5)
plt.show()

(convolvedData.T - interpolatedData.T).max()
```

There is a difference in the data. Now plot the convolved data vs the interpolated data for the fixed angles that we have taken data at:

```python
# First, find the indices for the fixed angles that we want
fixedExptAngles = np.array([20, 22.5, 25, 27.5, 30])
fixedTheoryAngIndices = np.zeros(fixedExptAngles.shape)
for i, angle in enumerate(fixedExptAngles):
    index = np.where(np.isclose(interpFixedAngles, angle))[0]
    fixedTheoryAngIndices[i] = index
    plt.figure(dpi = 276)
    plt.title(f'Fixed Angle: {angle}$^\circ$, $\sigma =$ {sigma:.2f}$^\circ$')
    plt.plot(interpSweptAngles, interpolatedData[:, index], label = 'Pure Theory')
    plt.plot(interpSweptAngles, convolvedData[:, index], label = 'Convolved Theory')
    dressFigure()

# Now get these slices and compare with
```
